```python
with open('throughput.txt', 'r') as f:
    throughput_table = [float(line.strip()) for line in f]

with open('throughput_with_isl.txt','r') as f:
    throughput_isl = [float(line.strip()) for line in f]
    
with open('throughput_with_coDld.txt','r') as f:
    throughput_coDld = [float(line.strip()) for line in f]
    
with open('latency.txt', 'r') as f:
    latency = [float(line.strip()) for line in f]
    
with open('latency_with_isl.txt', 'r') as f:
    latency_isl = [float(line.strip()) for line in f]
    
with open('latency_with_coDld.txt', 'r') as f:
    latency_coDld = [float(line.strip()) for line in f]   
    
with open('packet_droped_table.txt', 'r') as f:
    packet_droped_table = [float(line.strip()) for line in f]


with open('packet_droped_table_with_isl.txt', 'r') as f:
    packet_droped_isl = [float(line.strip()) for line in f]
    
with open('packet_droped_table_with_coDld.txt', 'r') as f:
    packet_droped_coDld = [float(line.strip()) for line in f]    
    
with open('satellite_queue_len.txt', 'r') as f:
    satellite_queue_len = [float(line.strip()) for line in f]
 
with open('satellite_queue_len_with_isl.txt', 'r') as f:
    satellite_queue_len_isl = [float(line.strip()) for line in f]
    
with open('ground_station_queue_len.txt', 'r') as f:
    ground_station_queue_len = [float(line.strip()) for line in f]
    
with open('ground_station_queue_len_with_isl.txt', 'r') as f:
    ground_station_queue_len_isl = [float(line.strip()) for line in f]
    
    


with open('throughput_with_umbra.txt','r') as f:
    throughput_umbra = [float(line.strip()) for line in f]
with open('latency_with_umbra.txt', 'r') as f:
    latency_umbra = [float(line.strip()) for line in f]
with open('packet_droped_table_with_umbra.txt', 'r') as f:
    packet_droped_umbra = [float(line.strip()) for line in f]
with open('satellite_queue_len_with_umbra.txt', 'r') as f:
    satellite_queue_len_umbra = [float(line.strip()) for line in f]
with open('ground_station_queue_len_with_umbra.txt', 'r') as f:
    ground_station_queue_len_umbra = [float(line.strip()) for line in f]

with open('throughput_with_flow.txt','r') as f:
    throughput_flow = [float(line.strip()) for line in f]
with open('latency_with_flow.txt', 'r') as f:
    latency_flow = [float(line.strip()) for line in f]
with open('packet_droped_table_with_flow.txt', 'r') as f:
    packet_droped_flow = [float(line.strip()) for line in f]
with open('satellite_queue_len_with_flow.txt', 'r') as f:
    satellite_queue_len_flow = [float(line.strip()) for line in f]
with open('ground_station_queue_len_with_flow.txt', 'r') as f:
    ground_station_queue_len_flow = [float(line.strip()) for line in f]
    
print("read success")
```

```python
plt.figure(dpi=1000)
plt.plot(throughput_table, label='Without ISL')
plt.plot(throughput_isl, label='With ISL')
plt.plot(throughput_umbra, label='With umbra')
plt.plot(throughput_flow, label='With flow')
plt.plot(throughput_coDld, label='With coDld')

plt.ylabel('Throughput')
plt.xlabel('Time')
plt.legend()
plt.show()
```
