```python
import matplotlib.pyplot as plt
```

```python
with open('throughput.txt', 'r') as f:
    throughput_table = [float(line.strip()) for line in f]
    
with open('throughput_with_isl.txt', 'r') as f:
    throughput_table_with_isl = [float(line.strip()) for line in f]
    
with open('throughput_with_umbra.txt', 'r') as f:
    throughput_with_umbra = [float(line.strip()) for line in f]
    
with open('throughput_with_max_flow_isl.txt', 'r') as f:
    throughput_with_max_flow_isl = [float(line.strip()) for line in f]
    
print("read success")
```

```python
plt.plot(throughput_table, label='Without ISL')
plt.plot(throughput_table_with_isl, label='With ISL')
plt.plot(throughput_with_umbra, label='With Umbra')
plt.plot(throughput_with_max_flow_isl,label='With Max Flow ISL')

plt.ylabel('Throughput')
plt.xlabel('Time')
plt.legend()
plt.show()
```

```python
prefix_sum = []
prefix_sum_with_isl = []
prefix_sum_with_umbra = []
prefix_sum_with_max_flow_isl = []
total=0
for i in throughput_table:
    total += i
    prefix_sum.append(total)
total=0
for i in throughput_table_with_isl:
    total += i
    prefix_sum_with_isl.append(total)
total=0
for i in throughput_with_umbra:
    total += i
    prefix_sum_with_umbra.append(total)
total=0
for i in throughput_with_max_flow_isl:
    total += i
    prefix_sum_with_max_flow_isl.append(total)
```

```python
plt.plot(prefix_sum, label='Without ISL')
plt.plot(prefix_sum_with_isl, label='With ISL')
plt.plot(prefix_sum_with_umbra, label='With Umbra')
plt.plot(prefix_sum_with_max_flow_isl,label='With Max Flow ISL')

plt.ylabel('Throughput')
plt.xlabel('Time')
plt.legend()
plt.show()
```
